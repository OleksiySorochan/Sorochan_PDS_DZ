```python
from math import sqrt

def math_func(x, P):
    Mx = 0   #  
    Dsum = 0
    for i in range(len(x)):
        Mx += x[i]*P[i]
        Dsum += x[i]*x[i]*P[i]
    Dx = Dsum - Mx*Mx # 
    sigma = sqrt(Dx) #  
    return Mx, Dx, sigma
```

###  X

```python
X = [1000, 1500, 2000]
Px = [0.5, 0.3, 0.2]
Mx, Dx, Sx = math_func(X, Px)
print(' ,  : ', Mx)
print('  ,  : ', Sx)
```

###  Y

```python
Y = [1000, 1500, 1750]
Py = [0.4, 0.4, 0.2]
My, Dy, Sy = math_func(Y, Py)
print(' ,  : ', My)
print('  ,  : ', Sy)
```
