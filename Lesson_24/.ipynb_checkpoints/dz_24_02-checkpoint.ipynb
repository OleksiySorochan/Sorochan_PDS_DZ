```python
from math import sqrt

def math_func(x, P):
    Mx = 0   # математичне сподівання
    Dsum = 0
    for i in range(len(x)):
        Mx += x[i]*P[i]
        Dsum += x[i]*x[i]*P[i]
    Dx = Dsum - Mx*Mx # Дисперсія
    sigma = sqrt(Dx) # квадратичне відхилення
    return Mx, Dx, sigma
```

### Набор X

```python
X = [1000, 1500, 2000]
Px = [0.5, 0.3, 0.2]
Mx, Dx, Sx = math_func(X, Px)
print('математичне сподівання, середнє значення: ', Mx)
print('середнє квадратичне відхилення, ступінь ризику: ', Sx)
```

### Набор Y

```python
Y = [1000, 1500, 1750]
Py = [0.4, 0.4, 0.2]
My, Dy, Sy = math_func(Y, Py)
print('математичне сподівання, середнє значення: ', My)
print('середнє квадратичне відхилення, ступінь ризику: ', Sy)
```

### Висновок:

Для двох наборів математичне сподівання , тобто середню значення прибутку і ймовірності однакове і складає: 1350. А ступінь ризику відрізнється для набора Х: 390, для набору Y: 300. Так, як ступінь ризику у набора Y меньше , то віддається перевага для виробництва і реалізації набору Y
