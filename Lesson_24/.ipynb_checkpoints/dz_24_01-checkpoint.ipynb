Для першого економіста ймовірності похибки і появи документа : 0.1 i 0.4;
для другого : 0.2 i 0.6

Ймовіність того, що навмання взятий документ із похибкою склав перший економіст: 

```python
0.1 * 0.4
```

Другий:

```python
0.2 * 0.6
```
