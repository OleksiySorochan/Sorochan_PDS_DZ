#### Спочатку спрощую таблицю, окремо для кожної акції

|X   | 10  | 5   | 2   |  1 | -5  |
|:---|:--- |:--- |:--- |:---|:--- |
|Px  | 0.1 | 0.3 | 0.2 |0.3 | 0.1 |

|Y   | 20  | 10   | 2   | -2 | -8  |
|:---|:--- |:---  |:--- |:---|:--- |
|Py  | 0.1 | 0.3  | 0.2 |0.3 | 0.1 |

```python
X = [10, 5, 2, 1, -5]
Px = [0.1, 0.3, 0.2, 0.3, 0.1]
Y = [20, 10, 2, -2, -8]
Py = [0.1, 0.3, 0.2, 0.3, 0.1]
```

```python
from math import factorial, pow, sqrt
```

```python
def my_faunc(x, P):
    Mx = 0   # математичне сподівання
    Mx2 = 0
    for i in range(len(x)):
        Mx += x[i]*P[i]
        Mx2 += pow(x[i], 2)*P[i]
    Dx = Mx2 - pow(Mx, 2)# Дисперсія
    sigma = sqrt(Dx) # квадратичне відхилення
    return round(Mx, 3), round(Dx, 3), round(sigma, 3)
```

```python
Mx, Dx, Sx = my_faunc(X, Px)
My, Dy, Sy = my_faunc(Y, Py)
print('Mx = ', Mx)
print('Dx = ', Dx)
print('Sx = ', Sx)
print('My = ', My)
print('Dy = ', Dy)
print('Sy = ', Sy)
```

### З таблиці створюю матрицю

```python
XY = [['y|x', 10, 5, 2, 1, -5],
      [20, 0.01, 0.03, 0.02, 0.03, 0.01],
      [10, 0.03, 0.09, 0.06, 0.9, 0.03],
      [2, 0.02, 0.06, 0.04, 0.06, 0.02],
      [-2, 0.03, 0.09, 0.06, 0.09, 0.03],
      [-8, 0.01, 0.03, 0.02, 0.03, 0.01]]
```

```python
# функція для обчислення математичного сподівання для двох акцій
def m_xy(arr):
    Mxy = 0
    for i in range(1, len(arr)):
        for j in range(1, len(arr[0])):
            Mxy += arr[i][0]*arr[0][j]*arr[i][j]
    return round(Mxy, 3)
```

```python
Mxy = m_xy(XY)
Mxy
```

```python
cov = round(Mxy - Mx*My, 3)
cov
```

### Коіфіцієнт кореляції

```python
rxy = round(cov / (Sx * Sy), 3)
rxy
```
