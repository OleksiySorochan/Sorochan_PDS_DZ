Створюємо множину із 30 елементів номерів днів

```python
all_day = set([i for i in range(1, 31)])
```

Функція для створення кастомної підмножини

```python
def create_costum_set(s, num):
    new_set = set()
    for i in range(num):
        new_set.add(list(s)[i])
    return new_set
```

Створюємо підмножину холодного, вітряного, і дощового дня із all_day

```python
HVD = create_costum_set(all_day, 1)
HVD
```

Створюємо підмножину вітряного, і дощового , 5 днів

```python
S = create_costum_set(all_day - HVD, 5 - len(HVD))
DV = S.union(HVD)
DV
```

Створюємо підмножину холодного і дощового , 3 дня

```python
S = create_costum_set(all_day - DV, 3 - len(HVD))
HD = S.union(HVD)
HD
```

Створюємо підмножину холодного і вітряного, 2 дня

```python
S = create_costum_set(all_day - DV.union(HD), 2 - len(HVD))
HV = S.union(HVD)
HV
```

Створюємо підмножину дощових днів , 12 днів

```python
a = all_day - DV.union(HD, HV)
S = create_costum_set(a, 12 - len(DV.union(HD)))
D = S.union(DV, HD)
D
```

Створюємо підмножину вітряних днів, 8 днів

```python
a = all_day - D.union(HV)
S = create_costum_set(a, 8 - len(DV.union(HV)))
V = S.union(DV, HV)
V
```

Створюємо підмножину холодних днів, 4 днів

```python
a = all_day - D.union(V)
S = create_costum_set(a, 4 - len(HV.union(HD)))
H = S.union(HV, HD)
H
```

```python
res = 30 - len(D.union(V, H))
print(f'{res} днів була гарна погода')
```
