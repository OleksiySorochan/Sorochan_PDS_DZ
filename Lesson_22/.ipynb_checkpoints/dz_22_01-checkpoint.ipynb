```python
all_st = set([i for i in range(1, 38)])
```

```python
def create_costum_set(s, num):
    new_set = set()
    for i in range(num):
        new_set.add(list(s)[i])
    return new_set
```

```python
super_st = create_costum_set(all_st, 4)
super_st   
```

```python
S = create_costum_set(all_st - super_st, 7 - len(super_st))
m_and_f = S.union(super_st)
m_and_f
```

```python
S = create_costum_set(all_st - m_and_f, 9 - len(super_st))
m_and_h = S.union(super_st)
m_and_h
```
