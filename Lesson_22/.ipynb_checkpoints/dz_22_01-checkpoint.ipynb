Множина зі всьома студентами. Цифри - це номер студента

```python
all_st = set([i for i in range(1, 38)])
```

Функція для створення кастомної підмножини 


```python
def create_costum_set(s, num):
    new_set = set()
    for i in range(num):
        new_set.add(list(s)[i])
    return new_set
```

Створення множини де студенти отримали відмінно зі трьох дісциплін. Вона є підмножнию all_st. Кількість елементів повинно бути 4

```python
super_st = create_costum_set(all_st, 4)
super_st   
```

 Математика та фізика 7 студентів, створюємо множину. Ця множина є підмножиною all_st, та множина super_st буде підмножиною цієї множини.

```python
S = create_costum_set(all_st - super_st, 7 - len(super_st))
m_and_f = S.union(super_st)
m_and_f
```

Математика та хімія - 9 студентів. Створюємо множину яка буде мати пересвчення із super_st, бути підмножиною all_st, та мати пересічення із m_and_f

```python
S = create_costum_set(all_st - m_and_f, 9 - len(super_st))
m_and_h = S.union(super_st)
m_and_h
```

Фізика та хімія - 6 студентів. Створюємо множину яка буде мати пересвчення із super_st, бути підмножиною all_st, та мати пересічення із m_and_f,  m_and_h

```python
S = create_costum_set(all_st - m_and_f - m_and_h , 6 - len(super_st))
f_and_h = S.union(super_st)
f_and_h
```

Математика - 15 студентів. Сворюємо множину яка буде мати пересічення із m_and_f та m_and_h, та бути підмножиною all_st

```python
a = all_st - m_and_f - m_and_h - f_and_h
S = create_costum_set(a , 15 - len(m_and_f.union(m_and_h)))
math = S.union(m_and_f, m_and_h)
math
```

Фізика 16 студентів

```python
a = all_st - math.union(f_and_h)
S = create_costum_set(a, 16 - len(m_and_f.union(f_and_h)))
physics = S.union(m_and_f, f_and_h)
physics
```

Хімія - 19 студентів

```python
a = all_st - math.union(physics)
S = create_costum_set(a, 19 - len(m_and_h.union(f_and_h)))
hh = S.union(m_and_h, f_and_h)
hh
```

Питання задачі полягає: скільки студентів отримали хочаб одну відміннц оцінку. Для цього потрібно вирахувати потужність трьох основних множин: математики, фізики і хіміі, але об'єднаних

```python
P = len(math.union(physics, hh))
print(f'{P} студента отримали хочаб одну відмінну оцінку')
```
