```python
all_day = set([i for i in range(1, 31)])
```

```python
def create_costum_set(s, num):
    new_set = set()
    for i in range(num):
        new_set.add(list(s)[i])
    return new_set
```

  , ,     all_day

```python
HVD = create_costum_set(all_day, 1)
HVD
```

  ,   , 5 

```python
S = create_costum_set(all_day - HVD, 5 - len(HVD))
DV = S.union(HVD)
DV
```

     , 3 

```python
S = create_costum_set(all_day - DV, 3 - len(HVD))
HD = S.union(HVD)
HD
```

    , 2 

```python
S = create_costum_set(all_day - DV.union(HD), 2 - len(HVD))
HV = S.union(HVD)
HV
```

    , 12 

```python
a = all_day - DV.union(HD, HV)
S = create_costum_set(a, 12 - len(DV.union(HD)))
D = S.union(DV, HD)
D
```

   , 8 

```python
a = all_day - D.union(HV)
S = create_costum_set(a, 8 - len(DV.union(HV)))
V = S.union(DV, HV)
V
```

   , 4 

```python
a = all_day - D.union(V)
S = create_costum_set(a, 4 - len(HV.union(HD)))
H = S.union(HV, HD)
H
```

```python
res = 30 - len(D.union(V, H))
print(f'{res}    ')
```
